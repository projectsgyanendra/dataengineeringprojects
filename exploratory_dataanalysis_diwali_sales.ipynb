```python
#total numbers of orders from top 10 countries
```

```python
df.columns
```

```python
#group by country and count orders 
top10_countries = df.groupby("State")["Orders"].sum().sort_values(ascending=False).head(10)  #calculates the total value
# convert to dataframe
top10_countries = top10_countries.reset_index()
# plot 
sns.barplot(x="State", y="Orders", data=top10_countries)
plt.title("Total Orders from Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Total sales per state") 
plt.xticks(rotation=70)  #xticks () is used to control the labels and position of ticks on the X-axis of a graph.
                         #Helps when labels overlap.
plt.show()
```

```python
#group by country and count orders 
top10_countries = df.groupby("State")["Orders"].count().sort_values(ascending=False).head(10) # counts the number of records
# convert to dataframe
top10_countries = top10_countries.reset_index()
# plot 
sns.barplot(x="State", y="Orders", data=top10_countries,hue = "State",palette="Set2",legend = False)
plt.title("Total Orders from Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Orders per state") 
plt.xticks(rotation=70)  #xticks () is used to control the labels and position of ticks on the X-axis of a graph.
                         #Helps when labels overlap.
plt.show()
```

From above graphs we can see that most of the orders are from uttar pradesh , maharashtra and karnataka respectively.

```python
#total amount/sales from top 10 states 
```

```python
df.groupby("State")["Amount"].sum().sort_values(ascending=False).head(10)
```

```python
# group by State and sum total sales/amount
top10_sales = df.groupby("State")["Amount"].sum().sort_values(ascending=False).head(10)

# convert to dataframe
top10_sales = top10_sales.reset_index()
sns.barplot(x="State", y="Amount", data=top10_sales,hue = "State" , palette="Set2" ,legend =  False)

plt.title("Total Sales from Top 10 States")
plt.xlabel("State")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=70)  # rotate labels for better readability

plt.show()
```

From above graphs we can see that total sales are from uttar pradesh , maharashtra and karnataka respectively.



Occupation

```python
#occupation

df.columns
```

```python
# group by Occupation and sum total Amount
# Group by Occupation and sum Amount
occupation_amount = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()

# Plot

sns.barplot(x='Occupation', y='Amount', data=occupation_amount, hue = 'Occupation' ,palette='Set2', legend = False)

plt.title("Top 10 Occupations by Total Amount")
plt.xlabel("Occupation")
plt.ylabel("Total Amount")
plt.xticks(rotation=70)  # rotate labels for readability

plt.show()
```

```python
# group by Occupation and sum total Amount
# Group by Occupation and sum Amount
occupation_amount = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()

# Plot
plt.figure(figsize=(20,15))  #figsize is used to set the size of the figure (plot) in inches. plt.figure(figsize=(width, height))
                          #width → width of the figure in inches , height → height of the figure in inches

sns.barplot(x='Occupation', y='Amount', data=occupation_amount, hue = 'Occupation' ,palette='Set2', legend = False)

plt.title("Top 10 Occupations by Total Amount")
plt.xlabel("Occupation")
plt.ylabel("Total Amount")
plt.xticks(rotation=0)  # rotate labels for readability

plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare , Aviation and Banking

Product Category

```python
#product category vs amount....
# group by Product Category and sum the Amount
category_amount = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,8))  # set figure size

sns.barplot(x='Product_Category', y='Amount',hue = 'Product_Category',data=category_amount, palette='Set2' , legend = False)

plt.title("Total Amount by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Amount")
plt.xticks(rotation=70)  # rotate labels for readability

plt.show()
```

from above graphs we can see that most of the sold products are from Food, Clothing .

```python
category_amount = df.groupby('Product_Category')['Amount'].count().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,2))  # set figure size

sns.barplot(x='Product_Category', y='Amount',hue = 'Product_Category',data=category_amount, palette='Set2' , legend = False)

plt.title("Total Amount by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total count")
plt.xticks(rotation=70)  # rotate labels for readability

plt.show()
```

from above graphs we can see that most of the order products are from clothing , food and electronics category
